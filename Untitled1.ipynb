```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
wine_dataset = pd.read_csv('winequality-red.csv')
```

```python
# number of rows & columns in the dataset
wine_dataset.shape
```

```python
# first 5 rows of the dataset
wine_dataset.head()
```

```python
# checking for missing values
wine_dataset.isnull().sum()
```

```python
# statistical measures of the dataset
wine_dataset.describe()
```

```python
# number of values for each quality
sns.catplot(x='quality', data = wine_dataset, kind = 'count')
```

```python
# Box plots for each chemical feature across different quality ratings
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
            'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

for feature in features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='quality', y=feature, data=wine_dataset)
    plt.title(f'{feature} by Wine Quality')
    plt.show()
```

```python
# volatile acidity vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y = 'volatile acidity', data = wine_dataset)
```

```python
# citric acid vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y = 'citric acid', data = wine_dataset)
```

```python
correlation = wine_dataset.corr()
```

```python
# constructing a heatmap to understand the correlation between the columns
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt = '.1f', annot = True, annot_kws={'size':8}, cmap = 'Blues')
```

```python
# Violin plot for alcohol content vs. wine quality
plt.figure(figsize=(10, 6))
sns.violinplot(x='quality', y='alcohol', data=wine_dataset)
plt.title('Alcohol Content vs. Wine Quality')
plt.show()

# Scatter plot for alcohol content vs. wine quality
plt.figure(figsize=(10, 6))
sns.scatterplot(x='alcohol', y='quality', data=wine_dataset)
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Scatter Plot of Alcohol Content vs. Wine Quality')
plt.show()
```

```python
# separate the data and Label
X = wine_dataset.drop('quality',axis=1)
```

```python
print(X)
```

```python
# Categorizing acidity levels
wine_dataset['acidity_level'] = pd.cut(wine_dataset['fixed acidity'], bins=[4, 6, 8, 16], labels=['low', 'medium', 'high'])
acidity_quality = wine_dataset.groupby('acidity_level')['quality'].mean()
print(acidity_quality)

# Bar plot for average wine quality by acidity level
acidity_quality.plot(kind='bar')
plt.ylabel('Average Quality')
plt.title('Average Wine Quality by Acidity Level')
plt.show()
```

```python
# Box plot for total sulfur dioxide vs. wine quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='total sulfur dioxide', data=wine_dataset)
plt.title('Total Sulfur Dioxide vs. Wine Quality')
plt.show()

# Scatter plot for total sulfur dioxide vs. wine quality
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total sulfur dioxide', y='quality', data=wine_dataset)
plt.xlabel('Total Sulfur Dioxide')
plt.ylabel('Quality')
plt.title('Scatter Plot of Total Sulfur Dioxide vs. Wine Quality')
plt.show()
```

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy : ', test_data_accuracy)
```

```python
# importing our models and metrics

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
svc = SVC(C=1.2, class_weight=None, gamma=0.9, kernel='rbf')
svc.fit(X_train, Y_train)
pred4 = svc.predict(X_test)

print(classification_report(pred4, Y_test), '\n')
print(sns.heatmap(confusion_matrix(pred4, Y_test), annot=True, cmap='coolwarm'), '\n')
print('Precision: ', precision_score(pred4, Y_test, average='macro'))
print('Accuracy: ', accuracy_score(pred4, Y_test))
print('Recall: ', recall_score(pred4, Y_test, average= 'macro'))
print('F1 Score: ', f1_score(pred4, Y_test, average= 'macro'))
```

LOGISTIC REGRESSION

```python
lg = LogisticRegression(random_state=10)
lg.fit(X_train, Y_train)
pred6 = lg.predict(X_test)

print(classification_report(pred6, Y_test), '\n')
print(sns.heatmap(confusion_matrix(pred6, Y_test), annot=True, cmap='coolwarm'), '\n')
print('Precision: ', precision_score(pred6, Y_test, average='macro'))
print('Accuracy: ', accuracy_score(pred6, Y_test))
print('Recall: ', recall_score(pred6, Y_test, average= 'macro'))
print('F1 Score: ', f1_score(pred6, Y_test, average= 'macro'))
```

```python
xgb = XGBClassifier(random_state=10)
xgb.fit(X_train, Y_train)
pred7 = xgb.predict(X_test)

print(classification_report(pred7, Y_test), '\n')
print(sns.heatmap(confusion_matrix(pred7, Y_test), annot=True, cmap='coolwarm'), '\n')
print('Precision: ', precision_score(pred7, Y_test, average='macro'))
print('Accuracy: ', accuracy_score(pred7, Y_test))
print('Recall: ', recall_score(pred7, Y_test, average= 'macro'))
print('F1 Score: ', f1_score(pred7, Y_test, average= 'macro'))
```
